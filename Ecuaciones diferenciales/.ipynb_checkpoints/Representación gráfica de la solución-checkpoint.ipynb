\begin{align}
\frac{d^2 x}{dt^2} +  \omega_{0}^2 x &= 0, \quad\mbox{donde} \quad \omega_{0} = \sqrt{\frac{k}{m}}\notag\\
\frac{d^2 \theta}{dt^2} + \omega_{0}^{2}\, \theta &= 0, \quad\mbox{donde}\quad \omega_{0}^2 = \frac{g}{l} 
\end{align} 

## Solución en el tiempo
El modelo del sistema `masa-resorte` está descrito por la siguiente **ecuación diferencial**:

\begin{equation}
\frac{d^2x}{dt^2} + \frac{k}{m}x = 0,
\end{equation}

cuya solución se escribe como 
\begin{equation}
x(t) = A \cos(\omega_{o} t) + B \sin(\omega_{o} t)
\end{equation}
Y su primera derivada (velocidad) sería 
\begin{equation}
v(t) = \frac{dx(t)}{dt} = \omega_{0}[- A \sin(\omega_{0} t) + B\cos(\omega_{0}t)]
\end{equation}

**¿Cómo se ven las gráficas de $x$ vs $t$ y $\frac{dx}{dt}$ vs $t$?** 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def pos(t,w0):
    A=0.5 # [m]
    B=0.5 # [m]
    return A*np.cos(w0*t)+B*np.sin(w0*t)
def vel(t,w0):
    A=0.5
    B=0.5
    return w0*(-A*np.sin(w0*t)+B*np.cos(w0*t))
```

```python
t=np.linspace(0,10,1000)
x=pos(t,1)
plt.plot(t,x,label='Posición [m]')
v=vel(t,1)
plt.plot(t,v,label='Velocidad [m/s]')
plt.xlabel('Tiempo [s]')
plt.legend()
plt.grid()
```

```python
#plt.legend?
'''
===============   =============
Location String   Location Code
===============   =============
'best'            0
'upper right'     1
'upper left'      2
'lower left'      3
'lower right'     4
'right'           5
'center left'     6
'center right'    7
'lower center'    8
'upper center'    9
'center'          10
===============   =============
'''
```

```python
W=np.linspace(0.1,2,10)
for w0 in W:
    x=pos(t,w0)
    v=vel(t,w0)
    plt.plot(t,x,label='Posición con $\omega_0='+str(round(w0,3))+'$')
    plt.plot(t,v,label='Velocidad con $\omega_0='+str(round(w0,3))+'$')
plt.legend(bbox_to_anchor=(1.01,0.5),loc='center left')
plt.grid()
```

```python
fig=plt.figure()
ax=fig.add_subplot(211)
for w0 in W:
    x=pos(t,w0)
    ax.plot(t,x,label='$\omega_0='+str(round(w0,3))+'$')
ax.set_title('Posición')
ax.legend(bbox_to_anchor=(1.01,0.5),loc=6)
ax.grid()
ax=fig.add_subplot(2,1,2)
for w0 in W:
    v=vel(t,w0)
    ax.plot(t,v,label='$\omega_0='+str(round(w0,3))+'$')
ax.set_title('Velocidad')
ax.legend(bbox_to_anchor=(1.01,0.5),loc=6)
ax.grid()
```

```python
fig=plt.figure()
gs=fig.add_gridspec(2,1,hspace=1)
ax=gs.subplots()
#fig,ax=plt.subplots(2,1)
fig.suptitle('Sistema masa-resorte')
for w0 in W:
    x=pos(t,w0)
    v=vel(t,w0)
    ax[0].plot(t,x,label='$\omega_0='+str(round(w0,3))+'$')
    ax[1].plot(t,v,label='$\omega_0='+str(round(w0,3))+'$')
ax[0].set(ylabel='m',title='Posición')
ax[1].set(xlabel='s',ylabel='m/s',title='Velocidad')
for axes in ax:
    axes.legend(bbox_to_anchor=(1.01,0.5),loc=6)
    axes.grid()
```

## Gráfico interactivo
La posición y velocidad para el sistema `masa-resorte` se escriben como: 
\begin{align}
x(t) &= x(0) \cos(\omega_{0} t) + \frac{\dot{x}(0)}{\omega_{0}} \sin(\omega_{0} t)\\
\dot{x}(t) &= -\omega_{0}x(0) \sin(\omega_{0} t) + \dot{x}(0)\cos(\omega_{0}t)
\end{align}

Importemos la clase `interact` e `interact_manual`

```python
from ipywidgets import *
```

**Actividad:** Crear un gráfico interactivo para variar la posición inicial $\dot x_0$ y la frecuencia fundamental $\omega_0$, para la velocidad del sistema `masa-resorte`.

```python
def pos(t,w0,x0=0.5,v0=0.5):
    return x0*np.cos(w0*t)+v0/w0*np.sin(w0*t)
def vel(t,w0,x0=0.5,v0=0.5):
    return -w0*x0*np.sin(w0*t)+v0*np.cos(w0*t)
```

```python
def fun_interact(w0,x0,v0):
    t=np.linspace(0,100,10000)
    for w_0 in w0:
        x=pos(t,w_0,x0,v0)
        v=vel(t,w_0,x0,v0)
        plt.plot(x,v,label='$\omega_0={}$'.format(w_0))
    plt.xlabel('Posición')
    plt.ylabel('Velocidad')
    plt.grid()
interact(fun_interact,w0=fixed([0.1,0.5,1]),x0=(-1,1,0.1),v0=(-1,1,0.1))
```

```python
@interact_manual(w0=fixed([0.1,0.5,1]),x0=(-1,1,0.1),v0=(-1,1,0.1))
def fun_interact(w0,x0,v0):
    t=np.linspace(0,100,10000)
    for w_0 in w0:
        x=pos(t,w_0,x0,v0)
        v=vel(t,w_0,x0,v0)
        plt.plot(x,v,label='$\omega_0={}$'.format(w_0))
    plt.xlabel('Posición')
    plt.ylabel('Velocidad')
    plt.grid()
```

**Animación del péndulo.**

La solución para el sistema reducido del `péndulo` es

\begin{align}
\theta(t) &= \theta(0) \cos(\omega_{0} t) + \frac{\dot{\theta}(0)}{\omega_{0}} \sin(\omega_{0} t)\\
\dot{\theta}(t) &= -\omega_{0}\theta(0) \sin(\omega_{0} t) + \dot{\theta}(0)\cos(\omega_{0}t).
\end{align}

Las posiciones $x$ y $y$ se obtienen de la descomposición entre la longitud y el ángulo:

\begin{align}
x &= l \sin(\theta), \quad
y = -l \cos(\theta) .
\end{align}
